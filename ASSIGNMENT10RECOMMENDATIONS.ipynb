```python
#Books along with its ratings
ratings=pd.DataFrame(books1.groupby('booktitle')['rating'].mean())
```

```python
#Added a column which counts the no of ratings for each book.
ratings["rating_num"]=books1.groupby('booktitle')['rating'].count()
```

```python
ratings
```

# collaboration matrix


```python
#steps to form collaboration matrix
user_books_df = books1.pivot_table(index='userid',
                                 columns='booktitle',
                                 values='rating')
```

```python
user_books_df
```

```python

user_books_df.index = books1.userid.unique()
```

```python
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
```

```python
user_books_df
```

# cosine similarity matrix

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python

user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books1.userid.unique()
user_sim_df.columns =books1.userid.unique()
```

```python
#Books based recommendations

ratings.sort_values("rating_num",ascending=False).head(10)  #we got Fahrenheit 451 has highest rating count
```

```python
#fetching the ratings for Fahrenheit 451
Fahrenheit_user_rating=user_books_df['Fahrenheit 451']
```

```python
#finding the books having correlation with Fahrenheit 451
similar_to_Fahrenheit=user_books_df.corrwith(Fahrenheit_user_rating)
```

```python
similar_to_Fahrenheit. head(10)
```

```python
#Making the dataframe with books having correlation with Fahrenheit 451 and its rating count

corr_fahrenheit=pd.DataFrame(similar_to_Fahrenheit,columns=['correlation'])
corr_fahrenheit.dropna(inplace=True)
corr_fahrenheit.head()
```

```python
corr_fahrenheit=corr_fahrenheit.join(ratings['rating_num'])
```
